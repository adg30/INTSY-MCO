## About the Proposed Dataset

This dataset lists 4,454 monsters and 34 different stats from various D&D 5e books. It works as a central file for monster info, gathering the numbers and details used to build creatures for the game. In D&D, Challenge Rating shows how tough a monster is in a fight. By building a tool that predicts this rating from basic stats, DMs can quickly check if their custom monsters are fair without doing the math themselves.

**Dataset Source:** [DnD 5e Bestiary](https://5e.tools/bestiary.html)

Here are the 34 stats that a monster might have. We sorted them into groups so you can see how they relate to each other. Each part includes a simple explanation of what it is, how it works in the data, and an example:

### General Information
1. **Name**: The unique title or identity of a creature. The primary label used to distinguish one entity from another in the bestiary. Example: `Tarrasque`
2. **Source**: The original publication or document where the information originated. The abbreviated code for the specific D&D book (e.g., MM'14) containing the stat block. Example: `MM'14 (Monster Manual 2014)`
3. **Page**: A specific surface of a sheet of paper within a book. The exact page number in the source book where the creature's full description is located. Example: `105 (Adult Brass Dragon)`
4. **Size**: The physical scale or dimensions of an object. A standardized category (Tiny, Small, Medium, Large, Huge, Gargantuan) that dictates the area a creature occupies in combat. Example: `Gargantuan (Ancient Amethyst Dragon)`
5. **Type**: A group of things sharing common traits. The biological or magical classification of the creature which determines how it is affected by certain spells and abilities. Example: `Dragon (Gem) (Adult Amethyst Dragon)`
6. **Alignment**: A position of agreement or alliance. A combination of moral (Good, Evil, Neutral) and ethical (Lawful, Chaotic, Neutral) codes that describe the creature's typical behavior. Example: `Lawful Evil (Adult Blue Dracolich)`

### Defensive Statistics
7. **AC (Armor Class)**: A measure of protection against physical or magical harm. The target numerical value an attacker must roll to successfully land a hit on the creature. Example: `19 (natural armor) (Adult Amethyst Dragon)`
8. **HP (Hit Points)**: The amount of vitality or endurance an entity possesses. The total damage a creature can take before it is defeated, usually expressed as an average and a dice formula. Example: `229 (17d12 + 119) (Adult Amethyst Dragon)`
9. **Speed**: The rate at which someone or something moves. The maximum distance in feet a creature can travel during its turn, covering various modes like walking, flying, or swimming. Example: `40 ft., Fly 80 ft. (hover), Swim 40 ft. (Adult Amethyst Dragon)`

### Ability Scores
10. **Strength**: Physical power and the ability to exert force. The score representing a creature's capacity for physical tasks, heavy lifting, and melee weapon damage. Example: `30 (Ancient Amonkhet Dragon)`
11. **Dexterity**: Agility, reflexes, and coordination. The score representing speed, balance, stealth, and accuracy with ranged or finesse weapons. Example: `20 (Demilich)`
12. **Constitution**: Physical health and stamina. The score representing the creature's endurance, which directly dictates its total Hit Points. Example: `30 (Tarrasque)`
13. **Intelligence**: The capacity for logic, reasoning, and memory. The score representing mental acuity and the ability to recall lore or perform complex calculations. Example: `20 (Adult Amethyst Dragon)`
14. **Wisdom**: Intuition, insight, and situational awareness. The score representing awareness of surroundings and the ability to resist mental influence. Example: `20 (Adult Moonstone Dragon)`
15. **Charisma**: Force of personality and social influence. The score representing a creature's magnetism, confidence, and internal magical power. Example: `30 (Archduke Zariel of Avernus)`

### Proficiencies and Resistances
16. **Saving Throws**: An attempt to avoid or mitigate a negative effect. Specific bonuses added to ability checks made to resist spells, poisons, or environmental hazards. Example: `Dex +7, Con +12, Wis +8, Cha +10 (Adult Amethyst Dragon)`
17. **Skills**: Expertise or proficiency in a specific task. Targeted bonuses for specific types of ability checks like Perception, Stealth, or History. Example: `Arcana +15, Perception +13, Stealth +7 (Adult Amethyst Dragon)`
18. **Damage Vulnerabilities**: A specific weakness that increases the effect of harm. Damage types that are doubled when applied to the creature. Example: `Fire (Amonkhet Mummy)`
19. **Damage Resistances**: The ability to withstand or dampen certain types of harm. Damage types that are reduced by half when applied to the creature. Example: `Cold, Radiant (Adult Crystal Dragon)`
20. **Damage Immunities**: Total protection from certain types of harm. Damage types that deal zero damage to the creature, regardless of the amount rolled. Example: `Lightning, Poison (Adult Blue Dracolich)`
21. **Condition Immunities**: Protection from specific physical or mental states. Status effects that cannot be forced upon the creature. Example: `Charmed, Exhaustion, Frightened, Paralyzed, Poisoned (Adult Blue Dracolich)`

### Senses and Interaction
22. **Senses**: Faculties for perceiving the physical world. Special methods of detection and the creature's baseline awareness (Passive Perception). Example: `Truesight 120 ft., passive Perception 21 (Ancient Dragon Turtle)`
23. **Languages**: Systems of communication. The specific spoken, written, or telepathic methods the creature uses to communicate with others. Example: `Common, Draconic, telepathy 120 ft. (Adult Amethyst Dragon)`

### Abilities and Combat
24. **Traits**: Inherent qualities or passive characteristics. Special passive rules that modify the creature's mechanics, such as "Magic Resistance." Example: `Legendary Resistance (3/Day) (Adult Blue Dracolich)`
25. **Actions**: The primary deeds or attacks performed on one's turn. The main things a creature does during its turn, such as weapon attacks or "Breath" weapons. Example: `Multiattack, Singularity Breath (Adult Amethyst Dragon)`
26. **Bonus Actions**: A secondary, faster task performed alongside a main action. Quick capabilities used in addition to a standard Action on the creature's turn. Example: `Psychic Step (Teleporting as a bonus action)`
27. **Reactions**: An immediate response to an event or trigger. Actions taken outside of the creature's turn in response to a specific event. Example: `Time Slip (Adult Time Dragon)`
28. **Legendary Actions**: Extraordinary feats performed by legendary beings. A pool of special actions a creature can take at the end of other creatures' turns. Example: `Explosive Crystal (Adult Amethyst Dragon)`
29. **Mythic Actions**: God-like powers used in the most climactic battles. High-tier actions available only after a boss's "Mythic Trait" is triggered. Example: `Armor of Storms (Ancient Dragon Turtle)`

### Environmental and World Effects
30. **Lair Actions**: Actions taken by a creature's home or environment. Environmental effects that occur on a specific initiative count when fighting a boss in its lair. Example: `Beguiling Whisper (Adult Amethyst Dragon)`
31. **Regional Effects**: Large-scale changes to the surrounding world. Passive magical changes to the environment within a set radius of a creature's lair. Example: `Crystal Profusion (Amethyst geodes forming nearby)`
32. **Environment**: The natural surroundings of a creature. The specific biomes where the creature is natively found. Example: `Arctic (Adult White Dragon)`
33. **Treasure**: Wealth, riches, or valuable items. The specific types of loot or magical items associated with the creature’s hoard. Example: `Relics (Adult Black Dragon)`

### Target Variable
34. **CR (Challenge Rating)**: A metric used to gauge the level of difficulty or danger a creature poses. The numerical value representing the creature's overall power level. This is the **dependent variable** that our model intends to predict. Example: `30 (The Tarrasque)`
