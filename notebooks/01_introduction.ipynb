**By**: Aaron Go, John Leomarc Alonzo, Sean Olores, and Sean Cardeño

**Dataset**: [DnD 5e Bestiary](https://5e.tools/bestiary.html) (downloadable)

**Motivation**: In Dungeons & Dragons 5th Edition, *Challenge Rating* (CR) is a numerical estimate of a monster's combat difficulty. It determines whether an encounter is trivially easy or potentially lethal for a given party level. Calculating it correctly involves cross-referencing offensive output, defensive resilience, and special abilities using a multi-step formula from the Dungeon Master's Guide. This is a process that is tedious and error-prone for custom *homebrew* content.

**Goal**: By the end of this project, our goal is to train and evaluate at least three supervised machine learning models (two classical models and one neural network) that can accurately predict the Challenge Rating of a D&D 5e monster given its raw stat block.

### **dataset description**

The dataset contains 4,454 monsters scraped from various 5th Edition sourcebooks. It includes 34 different features ranging from basic physical stats to complex traits like legendary actions. In D&D, **CR is the target variable/label** in that it represents the overall power of a creature.

Here is a breakdown of what each monster entry looks like:

#### **1. General Info**
- **Name**: Identity of the creature (e.g., `Tarrasque`).
- **Source/Page**: Where the monster was first printed (e.g., `MM'14`).
- **Size**: Physical scale (Tiny to Gargantuan).
- **Type**: Biological/magical category (Dragon, Undead, etc.).
- **Alignment**: Moral and ethical code (Lawful Good, etc.).

#### **2. Defense**
- **AC (Armor Class)**: How hard the monster is to hit.
- **HP (Hit Points)**: How much damage it can take before dying.
- **Speed**: Movement distance in feet. Also covers various modes like flying or swimming 

#### **3. Ability Scores**
- **STR, DEX, CON, INT, WIS, CHA**: The six core attributes representing physical and mental prowess.

#### **4. Resistances & Proficiencies**
- **Saving Throws/Skills**: Specific bonuses to rolls.
- **Vulnerabilities/Resistances/Immunities**: Special damage scaling logic.

#### **5. Combat Capabilities**
- **Traits**: Passive abilities (e.g., `Magic Resistance`).
- **Actions/Reactions**: Attacks and special moves.
- **Bonus Actions**: A secondary, faster task performed alongside a main action. 
- **Legendary/Mythic/Lair Actions/Regional Effects**: "Boss" level mechanics.

#### **6. Miscellaneous**
- **Senses**: Special methods of detection and the creature's baseline awareness.
- **Languages**: Systems of communication. 
- **Environment**: The natural surroundings of a creature.
- **Treasure**: The specific types of loot or magical items associated with the creature’s hoard.   

#### **Target Variable**
- **CR (Challenge Rating)**: The numerical value we want to predict (0 to 30). A metric used to gauge the level of difficulty or danger a creature poses.
